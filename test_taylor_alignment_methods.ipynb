## Import libraries

```python
import tomopy
from helperFunctions import MoviePlotter
from tomoDataClass import tomoData
from alignment_methods import sinogram_consistency_score
```

## Create Simulated Data

```python
#Set constants
numAngles = 200 #Number of projection angles
imageSize = 256 #Pixel size of one side of cubic array containing Shepp3D phantom
```

```python
#Create a simulated object
obj = tomopy.shepp3d(size=imageSize)
print(obj.shape)
print("Simulated Object")
MoviePlotter(obj) #Plots vertical slices through the phantom
```

```python
# Set projection angles
angles = tomopy.angles(nang=numAngles, ang1=1, ang2=360)

# create projection images/sinogram
projections = tomopy.project(obj, angles, pad=False)

print("Projections through object")
MoviePlotter(projections)
```

```python
tomo = tomoData(projections)
tomo.jitter(maxShift = 20)
tomo.add_noise()
print("Projections through Object with jitter and noise")
tomo.makeNotebookProjMovie()
```

## Align Data

```python
#Show bad reconstruction prior to alignment
tomo.reconstruct(algorithm='art')
print("\nBad reconstruction prior to alignment")
badRecon = tomo.recon.copy()
MoviePlotter(badRecon)
```

```python
# Baseline consistency score — before any alignment
# Baseline consistency score — before any alignment

tomo.reset_workingProjections(x_size=imageSize, y_size=imageSize) #You can adjust these for tighter cropping
tomo.normalize(isPhaseData=False)

combined_rmse_before, x_rmse_before, y_rmse_before, _, _ = sinogram_consistency_score(tomo, plot=True)
print(f"\nBaseline  →  Combined RMSE: {combined_rmse_before:.4f} px  |  X RMSE: {x_rmse_before:.4f} px  |  Y RMSE: {y_rmse_before:.4f} px")
print("++++++++++++++++++++++++++++++++")

tomo.cross_correlate_align(tolerance=0.001, maxShiftTolerance=0.5, max_iterations=15, stepRatio=1, yROI_Range=None, xROI_Range=None, isFull360=False, downsample=4, use_grad=True)
tomo.cross_correlate_align(tolerance=0, maxShiftTolerance=0, max_iterations=15, stepRatio=1, yROI_Range=None, xROI_Range=None, isFull360=False, downsample=2)
tomo.cross_correlate_align(tolerance=0, maxShiftTolerance=0, max_iterations=15, stepRatio=1, yROI_Range=None, xROI_Range=None, isFull360=False, downsample=1)
# tomo.vertical_mass_fluctuation_align(tolerance=0.1, max_iterations=15, y_range=None, sigma=2, smooth_sigma=3)
# tomo.PMA(max_iterations=5, tolerance=0, algorithm='art', crop_bottom_center_y=imageSize, crop_bottom_center_x=imageSize, standardize=False, levels=3, scale=2, iterations_per_level = [2,2,1])
tomo.make_updates_shift()

print("+++++++++++++++++++++++++++++++++")

# Post-alignment consistency score
combined_rmse_after, x_rmse_after, y_rmse_after, _, _ = sinogram_consistency_score(tomo, plot=True)

print(f"\nBaseline    →  Combined RMSE: {combined_rmse_before:.4f} px  |  X RMSE: {x_rmse_before:.4f} px  |  Y RMSE: {y_rmse_before:.4f} px")
print(f"After align →  Combined RMSE: {combined_rmse_after:.4f} px  |  X RMSE: {x_rmse_after:.4f} px  |  Y RMSE: {y_rmse_after:.4f} px")
print(f"\nImprovement:  ΔCombined = {combined_rmse_before - combined_rmse_after:.4f} px  |  ΔX = {x_rmse_before - x_rmse_after:.4f} px  |  ΔY = {y_rmse_before - y_rmse_after:.4f} px")
```
