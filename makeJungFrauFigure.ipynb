```python
import numpy as np
import matplotlib.pyplot as plt
import tifffile as tiff
from matplotlib.colors import LogNorm

def display_log_scaled_image(file_path):
    # Load the TIFF image
    img = tiff.imread(file_path).astype(np.float32)
    
    # Ensure there are no zero or negative values (LogNorm requires positive values)
    img[img <= 0.5] = 1e-4 # Replace 0s with smallest nonzero value

    # Plot the image with LogNorm color scaling (preserves actual data)
    fig, ax = plt.subplots(figsize=(8, 6))
    cax = ax.imshow(img, cmap="hot", norm=LogNorm(vmin=1e-4, vmax=img.max()))

    ax.set_xticks([])
    ax.set_yticks([])

    # Add a colorbar with the original scale
    cbar = fig.colorbar(cax, ax=ax)
    cbar.set_label("Pixel Intensity")

    plt.savefig("460JungFrauImage.png", dpi=300, bbox_inches='tight')
    plt.show()

# Example usage
display_log_scaled_image("ready_frame_data_xppl1026722_Run0460_01800.tiff")
```
